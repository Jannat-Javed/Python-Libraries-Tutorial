```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Line Plot

```python

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(12, 8))

# Plot
plt.subplot(2, 3, 1)  # 2 rows, 3 columns, 1st subplot
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
plt.subplot(2, 3, 2)  # 2 rows, 3 columns, 2nd subplot
plt.plot(x, y, marker='o', linestyle='--', color='r', label='Line 1')
plt.grid(True)
plt.legend()
plt.title('Custom Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

# Bar Plot


```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.subplot(2, 3, 3)  # 2 rows, 3 columns, 3rd subplot
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
plt.subplot(2, 3, 4)  # 2 rows, 3 columns, 4th subplot
plt.barh(categories, values, color='lightgreen')
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
```

# Histogram

```python

data = np.random.randn(1000)

plt.subplot(2, 3, 5)  # 2 rows, 3 columns, 5th subplot
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

# Pie Chart

```python
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']

plt.subplot(2, 3, 6)  # 2 rows, 3 columns, 6th subplot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Pie Chart')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8, 6))
x = np.linspace(1, 10, 50)
y = np.sin(x)

plt.scatter(x, y, color='red', marker='x')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
